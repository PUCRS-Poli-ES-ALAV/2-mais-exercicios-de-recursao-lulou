```python
 #Modele e implemente um método recursivo que calcule o fatorial de um número n passado como parâmetro.

# numero n!  = n x n-1 x (n-1)-1 x ... ate chegar a 0

def fatorial (n):
    if n == 0:
        return 1
    elif n > 0:
        return n * fatorial(n - 1)


print(fatorial(4))
```

```python
#Modele e implemente um método recursivo que calcule o somatório de um número n (passado como parâmetro) até 0.

# somatorio de n  = n + n-1 + (n-1)-1 x ... ate chegar a 0

def somatorio (n):
    if n != 0:
        if n > 0:
            return n + int(somatorio(n-1))
        else:
            return n + int(somatorio(n+1))
        
    elif n == 0:
        return 0

print(somatorio(4))
```

```python
#Modele e implemente um método recursivo que calcule o n-ésimo número da sequência de fibonacci.

#sempre soma-se o ultimo somado + o ultimo resultado

def fibonacci(n):
    if n == 1:
        return n
    elif n == 2:
        return n
    else:
        return fibonacci(n-1) + fibonacci(n-2)
    
print(fibonacci(9))
```

```python
#Modele e implemente um método recursivo que recebe um inteiro zero ou positivo e retorna um String com o número em binário.

def convBase2(n):
    if n > 0:
        print(n % 2)
        return convBase2(n // 2)
    else:
        return None

convBase2(25)

```

```python
#Modele e implemente um método recursivo que calcule o somatório dos números contidos em um ArrayList de inteiros, passado como parâmetro.

def calculaVetor(ar):
    if len(ar) > 1:
        ar[1] = ar[0] + ar[1]
        calculaVetor(ar[1:])
    else:
        return print(ar)

calculaVetor([1,2,3])
```

```python

#Modele e implemente um método recursivo para encontrar o maior elemento de um ArrayList.

def findBiggest(ar):
    if len(ar) == 1:
        return ar[0]

    if ar[0] > ar[1]:
        return findBiggest([ar[0]] + ar[2:])
    else:
        return findBiggest(ar[1:])

findBiggest([10,12,13])
```

```python
#Modele e implemente um método recursivo que calcule o somatório dos números inteiros entre os números k e j, passados como parâmetro.

# somatorio entre k e j: (k + j) + ((k-1)+(j-1)) + ... até k e j ficarem iguais -- contando que o k é o menor

def somatorio_entre(k, j):
    if k != j:
        return (k + j) + somatorio_entre((k+1),(j-1))
    else:
        return k
    

somatorio_entre(4, 6)
```

```python
#Modele e implemente um método recursivo que recebe um String em retorna true se este String for um palíndrome, false caso contrário.

# pega o primero caracter e coloca no final juntamente, pega o ultimo e coloca no começo

def eh_palindromo(s):
    if len(s) == 0 | len(s) == 1:
        return True
    elif s[0] != s[-1]:
        return False
    else:
        return eh_palindromo(s[1:-1])
    
eh_palindromo("ana")
```

```python
#Implemente um método recursivo para determinar se um string ocorre dentro de outro. ´boolean findSubStr(String str, String match)´

def findSubStr(string, match):
    if match == "":
        return True
   
    if len(string) < len(match):
        return False
    
    if string[:len(match)] == match:
        return True
    
    return findSubStr(string[1:], match)

print(findSubStr( "MOMOMOM", "MO"))
```

```python
#Implemente um métodos que recebe um String e retorna um ArrayList com todas as permutações deste String. ´ArrayList<String> permutations(String s)´

def permutations(s):
    if len(s) == 0:
        return [""]
    
    resultado = []
    
    for i in range(len(s)):
        letra_atual = s[i]
        resto = s[:i] + s[i+1:]
        
        perms_do_resto = permutations(resto)
        
        for perm in perms_do_resto:
            resultado.append(letra_atual + perm)
    
    return resultado
        
```
