```python
 #Modele e implemente um método recursivo que calcule o fatorial de um número n passado como parâmetro.

# numero n!  = n x n-1 x (n-1)-1 x ... ate chegar a 0

def fatorial (n):
    if n == 0:
        return 1
    elif n > 0:
        return n * fatorial(n - 1)


print(fatorial(4))
```

```python
#Modele e implemente um método recursivo que calcule o somatório de um número n (passado como parâmetro) até 0.

# somatorio de n  = n + n-1 + (n-1)-1 x ... ate chegar a 0

def somatorio (n):
    if n != 0:
        if n > 0:
            return n + int(somatorio(n-1))
        else:
            return n + int(somatorio(n+1))
        
    elif n == 0:
        return 0

print(somatorio(4))
```

```python
#Modele e implemente um método recursivo que calcule o n-ésimo número da sequência de fibonacci.

#sempre soma-se o ultimo somado + o ultimo resultado

def fibonacci(n):
    if n == 1:
        return n
    elif n == 2:
        return n
    else:
        return fibonacci(n-1) + fibonacci(n-2)
    
print(fibonacci(9))
```

```python
#Modele e implemente um método recursivo que calcule o somatório dos números inteiros entre os números k e j, passados como parâmetro.

# somatorio entre k e j: (k + j) + ((k-1)+(j-1)) + ... até k e j ficarem iguais -- contando que o k é o menor

def somatorio_entre(k, j):
    if k != j:
        return (k + j) + somatorio_entre((k+1),(j-1))
    else:
        return k
    

somatorio_entre(4, 6)
```

```python
#Modele e implemente um método recursivo que recebe um String em retorna true se este String for um palíndrome, false caso contrário.

# pega o primero caracter e coloca no final juntamente, pega o ultimo e coloca no começo

def eh_palindromo(s):
    if len(s) == 0 | len(s) == 1:
        return True
    elif s[0] != s[-1]:
        return False
    else:
        return eh_palindromo(s[1:-1])
    
eh_palindromo("ana")
```
